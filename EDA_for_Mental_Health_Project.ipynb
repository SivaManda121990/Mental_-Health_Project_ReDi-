```python
# Work Hours vs Stress Level

import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='Work Hours per Week',y='Stress Level',data=df)
plt.title('Work vs Stress Level')
plt.xlabel('Work Hours per week')
plt.ylabel('Stress Level (1-10)')
plt.show()
```

The average workweek is about 39.5 hours, with most people working between 30 and 50 hours.

Stress level varies a bit but stays mostly between 1 (low) and 3 (moderate/high).

Since the median stress level is 2, half of the people report moderate or lower stress.

Overall, the data suggests that most people work typical full-time hours with low to moderate stress, but stress doesn’t appear to spike drastically with more work hours based on this summary alone.



```python
# Stress Level among  the ones that work less or equal than 30 hours per work vs. people who work more then 30 hours

# Group 1: Work ≤ 30 hours
low_workload = df[df['Work Hours per Week'] <= 30]

# Group 2: Work > 30 hours
high_workload = df[df['Work Hours per Week'] > 30]


print("Average Stress Level (≤30 hours):", low_workload['Stress Level'].mean())
print("Average Stress Level (>30 hours):", high_workload['Stress Level'].mean())
```

Conclusion:
People working 30 hours or less have an average stress level of about 1.996.

Those working more than 30 hours have an average stress level of about 1.999.

The difference in average stress between the two groups is very small, suggesting that working more hours does not significantly increase stress levels in this dataset.

```python
import matplotlib.pyplot as plt

# Calculate mean stress levels
avg_stress_low = low_workload['Stress Level'].mean()
avg_stress_high = high_workload['Stress Level'].mean()

# Data for plotting
groups = ['≤30 Hours', '>30 Hours']
avg_stress = [avg_stress_low, avg_stress_high]

# Plotting
plt.figure(figsize=(6, 5))
bars = plt.bar(groups, avg_stress, color=['skyblue', 'salmon'])

# Add labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.02, f'{height:.2f}', ha='center', va='bottom')

plt.title('Average Stress Level by Work Hours Group')
plt.ylabel('Average Stress Level')
plt.ylim(0, max(avg_stress) + 1)
plt.tight_layout()
plt.show()
```

```python
# Relationship between daily screen time and self-reported happiness,stress,anxiety levels among individuals:

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. Correlation for numeric variables
correlations = df[['Screen Time per Day Hours', 'Happiness Score', 'Stress Level']].corr()
print("Correlation matrix:\n", correlations)

# 2. Scatter plots with regression lines for Happiness and Stress
plt.figure(figsize=(12, 5))

# Happiness vs Screen Time
plt.subplot(1, 2, 1)
sns.regplot(x='Screen Time per Day Hours', y='Happiness Score', data=df, scatter_kws={'alpha':0.4})
plt.title('Happiness vs Screen Time')
plt.xlabel('Screen Time per Day Hours')
plt.ylabel('Happiness Score')

# Stress Level vs Screen Time
plt.subplot(1, 2, 2)
sns.regplot(x='Screen Time per Day Hours', y='Stress Level', data=df, scatter_kws={'alpha':0.4}, color='orange')
plt.title('Stress vs Screen Time')
plt.xlabel('Screen Time per Day Hours')
plt.ylabel('Stress Level')

plt.tight_layout()
plt.show()

# 3. Boxplot: Screen Time by Mental Health Condition
plt.figure(figsize=(10, 6))
sns.boxplot(x='Mental Health Condition', y='Screen Time per Day Hours', data=df, palette='Set3')
plt.title('Screen Time by Mental Health Condition')
plt.xlabel('Mental Health Condition')
plt.ylabel('Screen Time per Day Hours')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Mean screen time by mental health condition
mean_screen_by_condition = df.groupby('Mental Health Condition')['Screen Time per Day Hours'].mean().sort_values()
print("\nMean Screen Time by Mental Health Condition:")
print(mean_screen_by_condition)
```

From this Correlation matrix:The correlation between screen time and:

Happiness is 0.017

Stress is 0.005.These values are very close to 0, which means:

There is no meaningful linear relationship between daily screen time and either happiness or stress level in my dataset.

Also, the correlation between happiness and stress is just 0.005, showing no significant association between those two either.In my data, screen time does not appear to strongly impact self-reported happiness or stress.



And from Mean Screen Time by Mental Health Condition:
Individuals with Depression (5.14 hrs) and Anxiety (5.12 hrs) report the highest average daily screen time.

Those with Bipolar disorder report the lowest screen time (5.00 hrs).

Individuals with no reported mental health condition (labeled 0) have a slightly lower screen time (5.08 hrs) than those with mental health conditions.

The overall differences are small (~0.1–0.13 hours or ~6–8 minutes) but consistent — people with mental health conditions tend to spend slightly more time on screens.

This may suggest that increased screen time is associated with mental health conditions like depression and anxiety.
